```python
seq_net = torch.nn.Sequential(torch.nn.Linear(10,20),torch.nn.ReLU(),torch.nn.Linear(20,10))
print(seq_net)
#  automatically generate the sequential layers
```

```python
# print(net)
```

```python
lossFunc = torch.nn.MSELoss()
optimizer = torch.optim.SGD(seq_net.parameters(),lr = lr,momentum = 0.9)
for i in range(epoch):
    output = seq_net(x)
    print(x.size())
    loss = lossFunc(output,x)
#     net.zero_grad()
    optimizer.zero_grad()
    loss.backward()
#     for val in net.parameters():
#         val.data.sub_(lr*val.grad.data)
#     print(i,'th loss',loss)
#     print(i,'th loss squeeze',loss.squeeze())
    optimizer.step()
    if (i+1) % 4 ==0:
        print(str(i+1)+'th loss',loss.squeeze())
```
